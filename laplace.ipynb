```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def laplace(x, u, b):
    return 1 / (2 * b) * np.exp(-np.abs(x - u) / b)
```

```python
reported_value = 100
epsilon = 0.5
```

```python
x = np.arange(95, 105, 0.1)
dist1 = laplace(x, reported_value, 1 / epsilon)
dist2 = laplace(x, reported_value + 1, 1 / epsilon)
plt.plot(x, dist1, label='distribution 1')
plt.plot(x, dist2, label='distribution 2')
plt.axvline(x=reported_value, c='black', dashes=(1, 2), label='reported value')
plt.legend()
```

```python
x = np.arange(-5,5,0.01)
dist1 = laplace(x, 0.26, 1 / 0.1)
dist2 = laplace(x, 0.04, 1 / 0.1)
plt.plot(x, dist1, label='distribution 1')
plt.plot(x, dist2, label='distribution 2')
plt.axvline(x=0.26, c='black', dashes=(1, 2), label='reported value')
plt.legend()
```

```python
x = np.arange(0,33,0.1)
dist1 = laplace(x, 28, 1 / 0.1)
dist2 = laplace(x, 26, 1 / 0.1)
dist3 = laplace(x, 22, 1 / 0.1)
dist4 = laplace(x, 15, 1 / 0.1)
dist5 = laplace(x, 8, 1 / 0.1)
dist6 = laplace(x, 5, 1 / 0.1)
plt.plot(x, dist1, label='Headache')
plt.plot(x, dist2, label='Cough')
plt.plot(x, dist3, label='Diabetes')
plt.plot(x, dist4, label='Asthma')
plt.plot(x, dist5, label='Dehydration')
plt.plot(x, dist6, label='Malaria')
plt.axvline(x=17.33, c='black', dashes=(1, 2), label='mean')
plt.legend()
```
